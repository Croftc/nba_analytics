## Train the Ensemble

```python
import numpy as np
import xgboost as xgb
from sklearn.calibration import CalibratedClassifierCV
from sklearn.utils import resample

class BootstrapCalibratedClassifier:
    def __init__(self, n_bootstrap_samples=8, base_model_params=None):
        self.n_bootstrap_samples = n_bootstrap_samples
        self.bootstrap_models = []
        self.weights_dicts = []
        
        # Default parameters for XGBoost classifier
        if base_model_params is None:
            self.base_model_params = {
                'tree_method': "hist",
                'sampling_method': 'gradient_based',
                'enable_categorical': True,
                'device': "cuda",
                'verbosity': 1,
                'max_depth': 7,
                'learning_rate': 0.022,
                'n_estimators': 300,
                'gamma': 7,
                'subsample': 1,
                'colsample_bytree': 0.8,
                'min_child_weight': 15,
                'reg_lambda': 0.3,
                'reg_alpha': 0.3
            }
        else:
            self.base_model_params = base_model_params

    def fit(self, X_train, y_train):
        for _ in range(self.n_bootstrap_samples):
            # Generate bootstrap sample
            X_train_sample, y_train_sample = resample(X_train, y_train, n_samples=len(X_train), replace=True)
            self.weights_dicts.append(X_train_sample['TEAM'].value_counts())
            
            # Create and fit the model
            model = xgb.XGBClassifier(**self.base_model_params)
            model.fit(X_train_sample, y_train_sample)
            self.bootstrap_models.append(model)

    def predict(self, X):
        # Aggregate predictions from bootstrap models
        preds = np.array([model.predict(X) for model in self.bootstrap_models])
        return np.round(np.mean(preds, axis=0)).astype(int)
    
    def predict_proba(self, X):
        # Aggregate predicted probabilities from bootstrap models
        proba = np.array([model.predict_proba(X) for model in self.bootstrap_models])
        # Average probabilities across all bootstrap models
        return np.mean(proba, axis=0)

    def score(self, X_test, y_test):
        # Evaluate each model and average the scores
        scores = [model.score(X_test, y_test) for model in self.bootstrap_models]
        return np.mean(scores)

# Usage
bootstrap_classifier = BootstrapCalibratedClassifier(n_bootstrap_samples=8)
bootstrap_classifier.fit(X_train, y_train)

# Evaluate on test set
print("Average Score:", bootstrap_classifier.score(X_test, y_test))

# Make predictions
predictions = bootstrap_classifier.predict(X_test)
```
